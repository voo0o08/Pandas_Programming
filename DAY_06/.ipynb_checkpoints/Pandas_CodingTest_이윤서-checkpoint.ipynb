# <hr>[문1] 아래 조건을 만족하는 Series를 생성하세요.
1) 데이터 : 1~50 범위에서 7의 배수 정수
2) 인덱스 : 'a~z' 범위에서 데이터 갯수 만큼의 알파벳
3) 타  입 : 부호가 없는 1바이트 정수
4) 이  름 : sevens
5) 속  성 : index, values, dtype, ndim, shape, size

```python
datas = list(range(7,50,7))
idx = map(chr, range(97,97+len(datas)))
sevens = pd.DataFrame(data=datas, index=idx, dtype = np.uint8)
sevens
```

```python
print("index\n=>", sevens.index, "\n\n")
print("value\n=>", sevens.values, "\n\n")
print("dtype\n=>", sevens.info(), "\n\n")
print("ndim\n=>", sevens.ndim, "\n\n")
print("shape\n=>", sevens.shape, "\n\n")
print("size\n=>", sevens.size, "\n\n")
```

# [문3] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stats_100701.xlsx
2) 년도별 인구가 가장 많은 지역과 가장 적은 지역
3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역
4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역은 지역

## 1) 데이터 : stats_100701.xlsx

```python
file2 = "../DATA/stats_100701.xlsx"
statDF = pd.read_excel(file2, index_col=[0], header=1)
statDF
```

```python
# 계는 필요없으니까  drop
statDF.drop("계", inplace=True)
```

```python
# 수도권도 drop
statDF.drop("수도권", inplace=True)
```

## 2) 년도별 인구가 가장 많은 지역과 가장 적은 지역

```python
# 2) 년도별 인구가 가장 많은 지역과 가장 적은 지역
def headTail(DF, year):
    print(year)
    print("가장 많은 지역", DF.sort_values(by=year, ascending=False).head(1).index)
    print("가장 적은 지역",DF.sort_values(by=year, ascending=False).tail(1).index)
    print("\n\n")

headTail(statDF, 2015)
headTail(statDF, 2016)
headTail(statDF, 2017)
headTail(statDF, 2018)
```

## 3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역

```python
# 3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역
sixDF = statDF.loc[["서울", "부산", "대구", "인천", "광주", "대전"]]
sixDF

headTail(sixDF, 2015)
headTail(sixDF, 2016)
headTail(sixDF, 2017)
headTail(sixDF, 2018)
```

## 4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역

```python
# 4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역
# "경기", "충북", "충남", "전북", "전남", "경북", "경남"

sevenDF = statDF.loc[["경기", "충북", "충남", "전북", "전남", "경북", "경남"]]
sevenDF

headTail(sevenDF, 2015)
headTail(sevenDF, 2016)
headTail(sevenDF, 2017)
headTail(sevenDF, 2018)
```

# <hr>[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx <br>
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정 <br>
3) 결측치 처리 => <br>
(가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거 <br>
(나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체 <br>
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
# 1) 데이터 : stock price.xlsx
file3 = "../DATA/stock price.xlsx"
# 2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
stDF = pd.read_excel(file3, index_col=[0])
stDF
```

## 3)- (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거

```python
stDF.dropna() # 하나라도 NaN일 때
```

```python
stDF.dropna(subset = "price") # price가 nan일 때
```

```python
stMask = (stDF["value"]>=1000)# value가 10000이하면 제거
stDF[stMask]
```

## (3)-(나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체

```python
# (3)-(나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
stDF1 = stDF.fillna(stDF.mean(numeric_only=True)) # 평균값으로 결측치 대체 -> 데이터 저장은 이것만 하겠습니다
```

```python
stDF1
```

```python
stDF.fillna(method="ffill") # 이전값으로 대체
```

```python
stDF.fillna(method="bfill") # 이후값으로 대체
```

## 4) value가 1000이하 이거나 price가 90000이상인 데이터


```python
mask = (stDF1["value"]<=1000) | (stDF1["price"] >= 90000)
stDF1[mask]
```
