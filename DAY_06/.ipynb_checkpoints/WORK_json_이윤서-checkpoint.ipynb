```python
def cityplot(xList, yList, cityName):
    plt.plot(xList, yList, linestyle='-')

    plt.title(cityName)
    plt.show()
```

```python
CityList = list(nobelDF2["properties.name_eng"])
```

```python
for idx in range(0, 44):
    if len((nobelDF2["geometry.coordinates"][idx])[0]) == 1:
        points1 = (nobelDF2["geometry.coordinates"][idx])[0][0]
        points2 = (nobelDF2["geometry.coordinates"][idx])[1][0]

        points = points1+points2

        xList1 = []
        yList1 = []
        for x, y in points1:
            xList1.append(x)
            yList1.append(y)
        xList2 = []
        yList2 = []
        for x, y in points2:
            xList2.append(x)
            yList2.append(y)
            plt.show()
        
        plt.plot(xList1, yList1, xList2, yList2, linestyle='-')
        plt.title(CityList[idx])
        
        
        #print(len(nobelDF2["geometry.coordinates"][idx]), end="\n\n") # 2
        #print((nobelDF2["geometry.coordinates"][idx])[0], end="\n\n") # list
        #print((nobelDF2["geometry.coordinates"][idx])[1], end="\n\n") # list
     
    else:
        points = (nobelDF2["geometry.coordinates"][idx])[0] 
        
        xList = []
        yList = []
        for x, y in points:
            xList.append(x)
            yList.append(y)
        cityplot(xList, yList, CityList[idx])
```

```python

plt.figure(figsize=(12,15))
for idx in range(0, 44):
    if len((nobelDF2["geometry.coordinates"][idx])[0]) == 1:
        points1 = (nobelDF2["geometry.coordinates"][idx])[0][0]
        points2 = (nobelDF2["geometry.coordinates"][idx])[1][0]

        points = points1+points2

        xList1 = []
        yList1 = []
        for x, y in points1:
            xList1.append(x)
            yList1.append(y)
        xList2 = []
        yList2 = []
        for x, y in points2:
            xList2.append(x)
            yList2.append(y)
        
        plt.plot(xList1, yList1, xList2, yList2, linestyle='-')
        #plt.title(CityList[idx])
        
        #print(len(nobelDF2["geometry.coordinates"][idx]), end="\n\n") # 2
        #print((nobelDF2["geometry.coordinates"][idx])[0], end="\n\n") # list
        #print((nobelDF2["geometry.coordinates"][idx])[1], end="\n\n") # list
     
    else:
        points = (nobelDF2["geometry.coordinates"][idx])[0] 
        
        xList = []
        yList = []
        for x, y in points:
            xList.append(x)
            yList.append(y)

        plt.plot(xList, yList, linestyle='-')

        plt.text(np.mean(xList), np.mean(yList), CityList[idx])

plt.show()
```
