# [데이터 원소값 변경 / 타입 변경 한꺼번 진행]

```python
mpgDF2 = pd.read_csv(file, header = None, sep="\s+", names =["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"] )
mpgDF2
```

```python
mpgDF2.info()
```

```python
# ?를 없애자
mpgDF2.horsepower.replace("?", 0.0, inplace=True)
mpgDF2["horsepower"].unique() # => "?" 없어진 거 확인 가능 
```

```python
# origin col 1-USA, 2-EU, 3-JPN
mpgDF2.origin.replace(1, "USA", inplace=True)
mpgDF2.origin.replace(2, "EU", inplace=True)
mpgDF2.origin.replace(3, "JPN", inplace=True)
# mpgDF2.origin.replace({1:"USA", 2:"EU", 3:"JPN"} inplace=True)
```

```python
mpgDF2.origin
```

```python
mpgDF2.origin.value_counts()
```

```python
mpgDF2.origin.nunique() # USA, JPN, EU라서 3개
```

```python
# horse col float으로 바꾸기 
# origin col category로 바꾸기
mpgDF2["horsepower"] = mpgDF2["horsepower"].astype("float64")
mpgDF2["origin"] = mpgDF2["origin"].astype("category")
mpgDF2.info()
```
