# <hr>데이터 형 수정하기 -> info()

```python
tiDF.info()
```

```python
tiDF["당시생존여부"] = tiDF["당시생존여부"].astype("bool")
tiDF["사회적 지위"] = tiDF["사회적 지위"].astype("category")
tiDF["성별"] = tiDF["성별"].astype("category")
# tiDF["나이"].astype("int")
tiDF["형제/배우자 수"] = tiDF["형제/배우자 수"].astype("int8")
tiDF["부모/자녀 수"] = tiDF["부모/자녀 수"].astype("int8")
tiDF["클래스"] = tiDF["클래스"].astype("category")
tiDF["남/여/어린이"] = tiDF["남/여/어린이"].astype("category")
tiDF["승선도시"] = tiDF["승선도시"].astype("category")
tiDF.info()
```

```python
# 나이에 이상치가 있다!!
tiDF["나이"].unique() # 소수인 값들은 반올림을 하고 nan을 없애주자
```

```python
tiDF["나이"].mean() # 결측치에 넣어줄 평균값 
```

```python
tiDF["나이"].fillna(tiDF["나이"].mean(), inplace=True)
```

```python
# 소수인 값들 반올림
tiDF["나이"] = tiDF["나이"].astype("int8")
```

```python
tiDF.info() # memory usage: 71.5+ KB ====> memory usage: 17.3 KB
```

# <hr>해석을 해보자!!

```python
tiDF2 = tiDF.copy()
```

```python
# 좌석 등급별 생존비율 
tiDF2.groupby(['클래스']).mean(numeric_only=True)

# .groupby(['Animal']).mean()
```

클래스를 그룹으로 나누어 각 클래스 별 생존여부를 보았을 때 클래스가 높을수록 생존여부가 높음을 알 수 있었다  
이런 젠장! 역시 돈이 최고다.....-명노아-<hr>

```python
tiDF2.groupby(['성별']).mean(numeric_only=True)
```

성별을 그룹으로 보았을 때 여성이 생존 여부가 더 높은 것을 알 수 있다.<hr>

```python
tiDF2.groupby(['남/여/어린이']).mean(numeric_only=True)
```

어린이는 자녀가 없음이 당연하여 평균적으로 보았을 때 최소 보호자가 1인은 있었음을 알 수 있고, 어린이가 혼자 살아남았을 확률은 낮기때문에 생존 비율을 보아 여성어른이 아이들을 데리고 탈출에 성공했음을 알 수 있다.

```python
tiDF2.groupby(['사회적 지위','클래스']).mean(numeric_only=True)
```

묶고 보니 사회적 지위와 클래스도 같은 의미를 가지는 col으로 하나는 없애도 해석에 문제가 없을 것같다.

```python
tiDF2.groupby(['당시생존여부']).mean(numeric_only=True)
```
