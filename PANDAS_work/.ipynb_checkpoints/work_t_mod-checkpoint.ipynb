# 1. 남북한 에너지

## 1-1. 북한 신재생 행 추가하기

```python
file = "../DATA/남북한발전전력량.xlsx"
```

```python
eDF = pd.read_excel(file, index_col=[0, 1])
```

북한 신재생 추가하기

## 1-2. 마지막 열(2017 이후)에 총계 추가

```python
eDF.info()
```

```python
eDF["총계"] = "-"
eDF
```

```python
for i in range(eDF.shape[0]):
    eDF.iloc[i] = eDF.iloc[i].replace("-", "0")
eDF = eDF.astype("int64")
for i in range(eDF.shape[0]):
    eDF["총계"][i] = eDF.iloc[i].sum()
eDF
```

# 2. 시도별 전출입지역

## 2-1. 전출 -> 전입 인구 분석

```python
file = "../DATA/시도별 전출입 인구수.xlsx"
```

```python
# 데이터 불러오기
popDF = pd.read_excel(file, index_col=[0, 1], header=[0, 1])
popDF
```

```python
for i in range(popDF.shape[0]):
    popDF.iloc[i] = popDF.iloc[i].replace("-", "0")
popDF = popDF.astype("int64")
popDF
```

```python
popDF.corr()
```

## 2-2. 서울과 제주로 가장 많이 이주한 지역

```python
# 서울로 가장 많이 이주한 지역 
# 합계 데이터 추가 
# 전입지역(서울/제주)만 index 삼아서 max 값을 찾고
# 전출지역이 어디인지 찾는다?!
```

```python
popDF2 = popDF.copy()
```

```python
popDF2["총계", "이동자수 (명)"] = 0
popDF2
```

```python
for i in range(popDF2.shape[0]):
    popDF2.iloc[i, -1] = popDF2.iloc[i].sum()
popDF2
```

```python
popDF2.index
```

```python
popDF2 = popDF2.reset_index()
```

```python
popDF2[("level_0", "")]
```

```python
########################################################## 서울 원본 : popDF2
# 전출 != "전국" 
outTF1 = (popDF2[("level_0", "")] != "전국")
# 전출 != "서울"
outTF2 = (popDF2[("level_0", "")] != "서울특별시")
# 전입 == "서울"
inTF1 = (popDF2[("level_1", "")] == "서울특별시")


SDF = popDF2[outTF1 & outTF2 & inTF1] # & inTF1 & inTF2
SDF
```

## 서울로 간 사람들...

```python
# 서울 전입을 제일 많이 한 곳
SDF.sort_values(by=("총계","이동자수 (명)"), ascending=False).head(1)
```

```python
########################################################## 제주 
# 전출 != "전국"
outTF1 = (popDF2[("level_0", "")] != "전국")

# 전출 != "제주"
outTF2 = (popDF2[("level_0", "")] != "제주특별자치도")

# 전입 == "제주"
inTF2 = (popDF2[("level_1", "")] == "제주특별자치도")

JejuDF = popDF2[outTF1 & outTF2 & inTF2] # & inTF1 & inTF2
```

## 제주로 간 사람들...

```python
# 제주 전입을 제일 많이 한 곳
JejuDF.sort_values(by=("총계","이동자수 (명)"), ascending=False).head(1)
```

# <hr>2-2 번 멀티 인덱스 그대로 가는 법

```python
popDF3 = popDF.copy()
popDF3
```

```python
popDF3["총계", "이동자수 (명)"] = 0

for i in range(popDF3.shape[0]):
    popDF3.iloc[i, -1] = popDF3.iloc[i].sum()
popDF3
```

## 서울로 간 사람들...

```python
popSoeul = popDF3.xs(key="서울특별시", level=1)
popSoeul = popSoeul.drop("전국")
popSoeul = popSoeul.drop("서울특별시")
```

```python
popSoeul.sort_values(by=("총계","이동자수 (명)"), ascending=False).head(1)
```

## 제주로 간 사람들...

```python
popJeju = popDF3.xs(key="제주특별자치도", level=1)
popJeju = popJeju.drop("전국")
popJeju = popJeju.drop("제주특별자치도")
```
