## [DataFrame 정렬]
- 종류 => 인덱스기반 정렬, 데이터/값 기반 정렬

```python
import pandas as pd
```

```python
filename = "../DATA/movies.csv"
```

- (3) 데이터 저장 => DataFrame
      * 파일 데이터 읽어서 저장 => pands.read_파일확장자() => return DF

```python
movieDF = pd.read_csv(filename)
movieDF
```

- 데이터 정보 확인 관련 메서드 

```python
# 데이터의 전체적인 구조 및 데이터 종류 요약
movieDF.info()
```

```python
# 컬럼별 통계적인 데이터 정보 -> 수치 데이터만 가능 / 그래도 궁금하면 include = "all"
twoDF = movieDF.describe() # DF
twoDF
```

```python
# 컬럼별 통계적인 데이터 정보 -> 수치 데이터만 가능 / 그래도 궁금하면 include = "all"
twoDF = movieDF.describe(include="all") # DF
twoDF
# unique, top(제일 많이 나온 애), freq(top이 몇번 나왔는지)가 새로 등장
```

```python
# 수치 컬럼(Rank, Year)의 최소 최대 평균 데이터 추출
twoDF.loc[["min", "max", "mean"]]
```

```python
pd.DataFrame([twoDF.Rank.values, twoDF.Year.values], columns=twoDF.index, index=twoDF.columns) # row, col 바꾸기!! 메서드는 아래서
```

- 메서드 : DF객체변수명.transepose()

- [실습] 다양한 정렬

```python
import numpy as np
from math import nan
data={'col1':['A','A','B',nan,'D','C'],
      'col2':[2,1,9,8,7,4],
      'col3':[0,1,9,4,2,3],
      'col4':['a','B','c','D','e','F'],}

dataDF=pd.DataFrame(data)
dataDF
```

- col1 컬럼을 기준으로 오름차순 정렬

```python
# nan = 결측치 값의 위치 설정 => na_position 파라미터 보통 제일 뒤로 감
dataDF.sort_values(by="col1", na_position="first")
```

```python
# col4 컬럼을 기주느로 오름차순 정렬
dataDF.sort_values(by="col4", key=lambda col:col.str.lower()) # col = 시리즈 col4 str이 왜 필요하노...ㅠㅠㅠㅠㅠㅠ
```
