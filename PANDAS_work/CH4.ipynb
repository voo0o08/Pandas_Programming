## 4.1.2 넘파이 ndarray로 DF 생성

```python
random_data = np.random.randint(1, 101, [3, 5])
random_data
```

```python
pd.DataFrame(random_data)
```

```python
# 리스트, 튜플처럼 규칙이 있고, 개수만 맞춰준다면 레이블 생성 가능
row_labels = ["Morning", "Afternoon", "Evening"]
col_labels = ["Mon", "Tue", "Wed", "Thu", "Fri"]
temporatures = pd.DataFrame(random_data, row_labels, col_labels)
temporatures
```

```python
# 시리즈와 마찬가지로 중복도 가능하지만 비추
row_labels = ["Morning", "Afternoon", "Morning"] # "Morning"이 두 개
col_labels = ["Mon", "Tue", "Wed", "Thu", "Fri"]
temporatures = pd.DataFrame(
    data = random_data, 
    index = row_labels, 
    columns= col_labels
)
temporatures
```

# 4.4 인덱스별 정렬

```python
nba = nba.sort_values(by = ["Team", "Salary"], ascending = [True, False])
nba # 정렬 영구적용 해버림 -> 행 인덱스를 통해 원본으로 돌려놓자
```

## 4.4.2 열 인덱스 기준으로 정렬

```python
nba = nba.sort_index(axis="columns")
nba
```

## 4.7.3 특정 열에서 값 추출

loc과 iloc 속성은 모두 두 번째 인수로 열을 받음 나도 열 받아 ㄹㅇ

```python
nba.loc["James Harden", ["Position", "Birthday"]]
```

```python
%%timeit
nba.at["Austin Rivers", "Birthday"]
# timeit은 셀에서 코드를 실행하고 실행하는데 걸리는 평균 시간을 계산
```
