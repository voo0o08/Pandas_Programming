```python
import pandas as pd
import numpy as np
```

```python
file = "../DATA/netflix.csv"
```

```python
netDF = pd.read_csv(file, parse_dates=["date_added"])
```

```python
netDF
```

### 1. 제한된 메모리 사용량과 사용성을 최대화하기 위해 데이터셋을 최적화하세요.

```python
netDF.info()
```

```python
netDF["type"] = netDF["type"].astype("category")
```

```python
netDF.info()
```

### 2. 제목이 "Limitless"인 모든 콘텐츠를 찾으세요.

```python
titleTF = (netDF["title"] == "Limitless")
netDF[titleTF]
```

### 3. "Robert Rodriguez"감독이 제작하고 유형이 "Movie"인 모든 콘텐츠를 찾으세요.

```python
TF = (netDF["director"] == "Robert Rodriguez") & (netDF["type"] == "Movie")
netDF[TF]
```

### 4. 등록된 날짜가 '2019-07-31' or 감독이 "Robert Altman"인 모든 콘텐츠를 찾으세요.

```python
TF = (netDF["date_added"] == "2019-07-31") | (netDF["director"] == "Robert Altman")
netDF[TF]
```

### 5. 감독이 "Orson Wells"/"Aditya Kripalani"/"Sam Raimi"인 모든 콘텐츠를 찾으세요.

```python
TF = netDF["director"].isin(["Orson Wells", "Aditya Kripalani", "Sam Raimi"])
netDF[TF]
```

### 6. 2019년 5월 1일과 2019년 6월 1일 사이에 등록된 모든 콘텐츠를 찾으세요.

```python
TF = netDF["date_added"].between("2019-05-01", "2019-06-01")
netDF[TF]
```

### 7.director 열에서 NaN값이 있는 모든 행을 삭제하세요.

```python
netDF.dropna(subset = ["director"])
```

### 8. 넷플릭스가 콘텐츠를 단 하나만 등록한 날짜를 식별하세요.

```python
netDF.drop_duplicates(subset = "date_added", keep = False)
```
