```python
import pandas as pd
```

```python
# 중첩 에피소드 데이터 정규화
# 각 에피소드 별도의 행으로 DF생성 / 메타 데이터 포함
```

```python
tv_shows_json = pd.read_json("../DATA/tv_shows.json")
tv_shows_json
```

```python
tv = pd.json_normalize(tv_shows_json["shows"], record_path="episodes", meta = ["show", "runtime", "network"])
tv
```

```python
# 2. 정규화된 데이터 셋을 각 프로그램 별로 하나씩 DF
xfiles = tv[tv["show"] == "The X-Files"]
lost = tv[tv["show"] == "Lost"]
slayer = tv[tv["show"] == "Buffy the Vampire Slayer"]
```
