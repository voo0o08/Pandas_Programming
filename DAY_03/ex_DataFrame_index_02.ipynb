## [DataFrame 인덱스 다루기]

- (1) 새로운 인덱스 설정 : 컬럼 ==> 인덱스 설정

### (1) 모듈 불러오기

```python
import pandas as pd
```

### (2) 데이터 준비

```python
# Dict type의 데이터를 준비한다
exam = {'이름' : [ '마징가', '배트맨', '홍길동','임꺽정'],
'국어' : [ 98, 79, 99, 100],
'미술' : [ 70, 99, 87, 80],
'수학' : [ 77, 70, 98, 72],
'체육' : [ 77, 93, 69, 91]}
```

### (3) 데이터 저장

```python
examDF = pd.DataFrame(exam)
examDF
```

### (4) 데이터 확인

- 행 데이터 추가

```python
# 마징가 98 50 60 77
examDF.loc[4]=['마징가',98,50,60,77]
examDF
```

```python
examDF.loc[5]=['마징가', 98,87,70,91]
examDF
```

- (5) 새로운 인덱스 설정 - 방법3 : 멀티인덱스 제어 DF.set_index([col1, col2...])

```python
# 이름, 국어, 수학을 인덱스 설정
examDF1 = examDF.set_index(["이름", "국어", "수학"]) 
examDF1 # 출력 중 빈공간은 중복이라서 생략된 것, 없는게 아님!!
```

```python
examDF1.index # DF1의 멀티인덱스 확인
```

- 행데이터 선택 : DF객체변수명.loc[행레이블] / DF객체변수명.iloc[위치인덱스]

```python
examDF1.loc[('마징가',  98, 77)]
```

```python
# 근데 국어가 98인 애들만 뽑고 싶으면 어떡해..? 나 울어...?
# 이미 멀티인덱스 상태라 examDF1.loc[(98)] 불가능ㅠㅠ
```

- 멀티인덱스의 인덱스의 일부분을 기준으로 행 선택 : xs()
    * ex) "마징가" 98 77 => level : 0 1 2로 구분가능

```python
examDF1.xs("마징가", level = 0) # 멀티인덱스 중 "마징가" 기준
```

```python
examDF1.xs(98, level = 1) # 멀티인덱스 중 "국어" 기준
```

```python
examDF1.xs(70, level = 2) # 멀티인덱스 중 "수학" 기준
```

- ("마징가", 98, 77), ("임꺽정", 100, 72) 2개 행을 한꺼번에 가져오기

```python
examDF1.loc[[("마징가", 98, 77), ("임꺽정", 100, 72)]] # 2개 가져오고 싶으면 리스트 안에 넣기!!
# 1개할 때도 [("마징가", 98, 77)] -> 시리즈 / [[("마징가", 98, 77)]] -> 데이터프레임
```

- (5) 인덱스 재정렬 : 행/열 인덱스를 원하는 순서대로 재배치 + 존재하지 않는 행/열 추가
    * DF.index()

```python
examDF # reindex를 사용해보자~
```

```python
print(examDF.index, examDF.columns, sep = "\n\n")
```

['이름', '국어', '미술', '수학', '체육'] -> ["체육", "미술", "국어", "수학", "이름"]<br>
기본값이 행이니 axis를 바꾸자

```python
examDF.reindex(["체육", "미술", "국어", "수학", "이름"])
```

```python
examDF.reindex(["체육", "미술", "국어", "수학", "이름"], axis = 1)
```

```python
examDF.reindex(["체육", "국어", "이름", "총계"], axis=1) # NaN이 마음에 안들면 fill_value로 0주거나 하면 됨
```

```python
examDF.reindex(["체육", "국어", "이름", "총계"], axis=1, fill_value=0)
```

```python
examDF.reindex([0, 2, 4, 1, 3, 5], axis=0) # axis=0이 default라 안써도 상관x
```

axis=0, axis=1 다 쓰고 싶다면?

```python
examDF.reindex(index=[0, 2, 4, 1, 3, 5], columns=["체육", "국어", "이름", "총계"], fill_value=100)
```
