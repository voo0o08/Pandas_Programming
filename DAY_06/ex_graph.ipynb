# [데이터 시각화 맛보기]

## <hr>(1) 모듈 로딩

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

## <hr>(2) 데이터 준비

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = range(0, 100, 10)
y = [x**2 for x in x]
```

## <hr>(3) 그래프 그리기

- [기본] : 라인/선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적이고 순서가 있는 데이터를 시각화에 사용할 경우 사용

```python
# 반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨 
ret = plt.plot(x, y)
type(ret) , len(ret), type(ret[0]) # 그림의 객체가 ret에 담기는 것 
```

```python
ret[0].get_xdata()
```

- 라인/선 그래프에서 형식 즉 포맷지정 => str 타입의 포맷 설정
    * 색상-형태-선스타일 ==> 'ro--'(red, circle, dot-line)
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, c - cyan, m - magenta, y - yellow, k - black, w - white
    * 형태 : o - circle, ^ - triangle, s - square, v - pentagram, < - left triangle, > - right triangle, 8 - octagon, p - pentagram, * - star, h - hexagon, + - plus, x - x, D - diamond, d - thin_diamond, | - vline, _ - hline
    * 선스타일 : - - solid line, -- - dashed line, -. - dash-dot line, : - dot-dash line
    * 포맷 설정 방법 : plt.plot(x,y,'r--')

```python
ret = plt.plot(x, y, "ro")
type(ret) , len(ret), type(ret[0]) # 그림의 객체가 ret에 담기는 것 
```

```python
ret = plt.plot(x, y, "r--")
type(ret) , len(ret), type(ret[0]) # 그림의 객체가 ret에 담기는 것 
```

```python
ret = plt.plot(x, y, "g^")
type(ret) , len(ret), type(ret[0]) # 그림의 객체가 ret에 담기는 것 
```

```python
x2 = [x//2 for x in x] 
y2 = [x**2.1 for x in x]
ret = plt.plot(x, y,"go" , x2, y2, "y^")

plt.show()
```
