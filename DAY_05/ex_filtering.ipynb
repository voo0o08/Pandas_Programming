### [데이터 필터링]
- 원하는 데이터를 추출하기 위해서 사용하는 방법
- 원하는 데이터 걸러내겠다는 의미
- 다양한 조건 검사 : 같다== / 같지않다!= / 크다> / 크거나 같다>=

<hr>(1) 모듈 로딩

```python
import numpy as np
import pandas as pd
```

```python
file = "../DATA/employees.csv"
```

<hr>(2) 데이터 저장

```python
empDF = pd.read_csv(file)
```

```python
empDF["Start Date"] = pd.to_datetime(empDF["Start Date"])
```

```python
empDF.info()
```

<hr>(5) 데이터 필터링

employees 데이터에서 이름이 Maria인 데이터만 추출

```python
empDF
```

```python
mariaMask = (empDF["First Name"] == "Maria") # bool값으로 구성된 데이터 ==> 불린 인덱싱(Boolean Indexing)
mariaMask
```

```python
empDF[mariaMask]
```

[문제] employees 데이터에서 연봉이 50000이상인 데이터 추출

```python
SalaryMask = (empDF["Salary"] >= 50000) # bool값으로 구성된 데이터 ==> 불린 인덱싱(Boolean Indexing)
empDF[SalaryMask]
```

[문제] 성별이 남자이고, 연봉이 100,000이상인 사람

```python
Mask = (empDF["Salary"] >= 148000) & (empDF["Gender"] == "Male") # bool값으로 구성된 데이터 ==> 불린 인덱싱(Boolean Indexing)
empDF[Mask]
```

- 여러 개의 조건이 모두 True여야만 하는 조건 => &
    * 문법 : 조건1 & 조건2 & 조건3 & ... & 조건n

- 여러 개의 조건 중 1개 이상 True여야만 하는 조건 => |
    * 문법 : 조건1 | 조건2 | 조건3 | ... | 조건n
    * 연산자 : |파이프 
    * 결과 : T/F

```python
Mask = (empDF["Start Date"] <= "1993-01-01") | (empDF["Mgmt"] == False) # bool값으로 구성된 데이터 ==> 불린 인덱싱(Boolean Indexing)
empDF[Mask]
```
