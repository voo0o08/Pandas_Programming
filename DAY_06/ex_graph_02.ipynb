### plt.text 기능들
- 정렬 제어(수직)
    * "center" : 텍스트 상자의 가운데와 관련
    * "top" : 텍스트 상자의 위쪽과 관련
    * "bottom" : 텍스트 상자의 아래쪽과 관련
    * "baseline" : 텍스트 기준선과 관련
- 정렬 제어(수평)
    * "center" : 텍스트 상자의 가운데와 관련
    * "left" : 텍스트 상자의 왼쪽과 관련
    * "right" : 텍스트 상자의 오른쪽과 관련
    * "baseline" : 텍스트 기준선과 관련
- 경계상자 제어 -> bbox = {} 에 파라미터 입력
    * "facecolor" : 상자에 사용하는 컬러
    * "edgecolor" : 상자 엣지 컬러
    * "alpha" : 상자와 배경을 혼합하기 위해 투명도를 설정할 때 사용
    * "boxstyle" : "round" | "square"
    * "pad" : "boxstyle"=="square" -> 텍스트와 상자 사이를 채울 용량 

```python
# 경계 상자 제어 
box = {"facecolor" : ".75", # 숫자 여러개 넣으면 r,g,b,alpha이지만 1개면 흑백의 비율
      "edgecolor" : 'y',
      "boxstyle" : "round"}

plt.text(-0.5, -0.25, "Brackmard minimum", bbox=box)

plt.plot(X, Y, c='k') # c=black
plt.show()
```

## 화살표 추가 plt.annotate("text", ha, va, xytext, xy, arrowprops)

```python
plt.annotate("text", 
             ha="center", va = "bottom", 
             xytext = (-1.5, 3), # 텍스트의 위치
             xy = (0.75, -2.7), # 화살표의 목적지
             arrowprops = {"facecolor": "0", "shrink": 0.05})
# shrink는 막대 길이 0~1
plt.plot(X, Y, c='k') # c=black
plt.show()
```

## 범례 추가 plt.legend()
- "loc" : 범례 위치다. 기본값은 'best' 로서 자동으로 배치한다. 다른 유효한 값은 'upper left' 'lower left', '1ower right' 'right'. 'center left', r'다
'center r
- i- ght'
'shadow ': True 또는 False 중 하나이며, 법례에 음영 효과를 <br>렌- 더링한다.
'fancybox': True 또는 또는 False ralse 중 하나이며. 범레에 로서리가 근 사각형상
자<br>를-  렌더링한다.
'title': 파라미터로 전달한 제목을 범<br>레- 에 렌ol한다.
'nc01': 범례의 열 개수가 전달한 값이 되도록 강제로 지정한다.

```python
X = np.linspace(0, 6, 1024)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.xlabel("X")
plt.ylabel("Y")

plt.plot(X, Y1, c='k', lw = 3., label = "sin(X)") # lw = 선 두께
plt.plot(X, Y2, c='k', lw = 3., ls = "--", label = "sin(X)") # ls = 라인 스타일 

plt.legend() # plot에서 label 설정 후 넣어주면 됨 
plt.show()
```
