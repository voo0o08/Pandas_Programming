## 5.1.1 astype 메서드를 사용하여 데이터 유형 변환

```python
empDF["Mgmt"] = empDF["Mgmt"].astype(bool)
```

```python
empDF.info() # 메모리 줄어듦 
```

```python
empDF["Salary"] = empDF["Salary"].fillna(0).astype(int)
```

```python
empDF.info()
```

```python
# 범주형 데이터 : 성별 = 카테고리 데이터 
empDF.nunique() 
```

```python
empDF["Gender"].unique() # nunique하면 nan을 포함 안함 
```

```python
empDF["Gender"] = empDF["Gender"].astype("category")
```

```python
empDF
```

```python
empDF["Team"] = empDF["Team"].astype("category")
```

```python
empDF.info() # 메모리가 아주 많이 줄었음을 확인할 수 있음 
```

# 단일 조건으로 필터링

```python
# 이미 bool type인 Mgmt는 Mask로 사용 가능 
# 기업의 모든 관리자 검색 Mgmt
empDF[empDF["Mgmt"]]
```

# 5.4 조건별 필터링

## 5.4.1 isin 메서드

```python
all_star_teams = ["Sales", "Legal", "Marketing"]
on_all_star_teams = empDF["Team"].isin(all_star_teams)
empDF[on_all_star_teams]
```

## 5.4.2 between 메서드

```python
between_and = empDF["Salary"].between(80000, 90000)
empDF[between_and]
```
