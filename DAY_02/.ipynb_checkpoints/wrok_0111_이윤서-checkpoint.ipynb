## (6) 컬럼별 데이터의 속성 즉, index, values, shape, ndim, size, dtype 정보 출력

```python
print(movieDF.columns)
colList = movieDF.columns
```

```python
for colName in colList:
    print(f"--------------------------------------------------{colName}의 속성 정보--------------------------------------------------")
    print(f"index : {cols[colName].index}")
    print(f"vaules : {cols[colName].values}")
    print(f"shape : {cols[colName].shape}")
    print(f"ndim : {cols[colName].ndim}")
    print(f"size : {cols[colName].size}")
    print(f"dtype : {cols[colName].dtypes}")
    print()
    # cols[colName]
```

## (7) 컬럼별 인덱스를 변경하기

```python
# cols -> col별 딕셔너리
# cols["Rank"].index = range(3, 3+782)
cols["Rank"].index = [str(i)+"*" if i%2 else i for i in range(3, 3+782)] # 한 index에 다른 type이 섞이기는 OK
cols["Rank"]=cols["Rank"].rename(index = {"3*" : "a"}) # 전체 말고 하나만 바꾸고 싶다면 rename 사용 / 저장은 안되고 return값 받아야함
cols["Rank"]
```

## (8) 컬럼별 데이터 즉 값 변경하기

```python
# 랭킹 순위를 100위부터
rowN = cols["Rank"].shape[0] # 요소의 개수
cols["Rank"][:] = [i for i in range(100, 100+rowN)] # 웃기는 짬뽕이다... index는 '.index = 어쩌구'로 바꾸고 값은 ".values = 어쩌구 안됨...?"
cols["Rank"]
```

```python
# 랭킹 순위를 100위를 0위로
rowN = cols["Rank"].shape[0] # 요소의 개수
cols["Rank"].iloc[0] = 0 # datatype 시리즈일 때도 가능!! iloc이 꼭 dataframe에만 먹히는 건 아님!! 왜지...
cols["Rank"]["5*"] = "여기다 여기"
cols["Rank"]
```

## (9) DataFrame의 원소를 모두 출력하기 단, 2가지 방법 사용

```python
movieDF
```

```python
# iloc 사용한 ver
rowN, colN = movieDF.shape
for col in range(colN):
    print(f"-------{movieDF.columns[col]}-------")
    for row in range(rowN):
        print(movieDF.iloc[row, col], end = " ")
    print("\n")
```

```python
# loc 사용한 ver -> loc은 이름만 받는다~
rowN, colN = movieDF.shape
for colName in movieDF.columns:
    print(f"-------{colName}-------")
    for rowName in movieDF.index:
        print(movieDF.loc[rowName, colName], end = " ")
    print("\n")
```
