## 1-2. 마지막 열(2017 이후)에 총계 추가

```python
eDF.info()
```

```python
eDF["총계"] = "-"
eDF
```

```python
for i in range(eDF.shape[0]):
    eDF.iloc[i] = eDF.iloc[i].replace("-", "0")
eDF = eDF.astype("int64")
for i in range(eDF.shape[0]):
    eDF["총계"][i] = eDF.iloc[i].sum()
eDF
```

# 2. 시도별 전출입지역

## 2-1. 전출 -> 전입 인구 분석

```python
file = "../DATA/시도별 전출입 인구수.xlsx"
```

```python
# 데이터 불러오기
popDF = pd.read_excel(file, index_col=[0, 1], header=[0, 1])
popDF
```

## 2-2. 서울과 제주로 가장 많이 이주한 지역

```python
# 서울로 가장 많이 이주한 지역 
# 합계 데이터 추가 
# 전입지역(서울/제주)만 index 삼아서 max 값을 찾고
# 전출지역이 어디인지 찾는다?!
```

```python
popDF2 = popDF.copy()
popDF2
```

```python
new_index = pd.MultiIndex.from_tuples(["총계"]) # , "이동자수"
new_col = pd.DataFrame(index=popDF2.index, columns=new_index).fillna("-")

result = pd.concat([popDF2, new_col], axis = 1)
result
```

```python
for i in range(result.shape[0]):
    result.iloc[i] = result.iloc[i].replace("-", "0")
result = result.astype("int64")

for i in range(result.shape[0]):
    result.iloc[i, -1] = result.iloc[i].sum()
result

```

# 2-2. [다시]

```python
# 데이터 불러오기
popDF = pd.read_excel(file, index_col=[0, 1], header=None, skiprows=2)
popDF
```

```python
new_col = pd.DataFrame(index=popDF.index, columns=["총계"]).fillna("-")

result = pd.concat([popDF, new_col], axis = 1)
result
```

```python
for i in range(result.shape[0]):
    result.iloc[i] = result.iloc[i].replace("-", "0")
result = result.astype("int64")

for i in range(result.shape[0]):
    result.iloc[i, -1] = result.iloc[i].sum()
result = result.reset_index()
```

```python
########################################################## 서울 
# 전출 != "전국"
outTF1 = (result[0] != "전국")
# 전출 != "서울"
outTF2 = (result[0] != "서울특별시")
# 전입 == "서울"
inTF1 = (result[1] == "서울특별시")


SDF = result[outTF1 & outTF2 & inTF1] # & inTF1 & inTF2
SDF
# movies[movies["Studio"] == "Universal"]
```

```python
# 서울 전입을 제일 많이 한 곳
SDF.sort_values(by="총계", ascending=False).head(1)[0]
```

```python
########################################################## 제주 
# 전출 != "전국"
outTF1 = (result[0] != "전국")

# 전출 != "제주"
outTF2 = (result[0] != "제주특별자치도")

# 전입 == "제주"
inTF2 = (result[1] == "제주특별자치도")

JejuDF = result[outTF1 & outTF2 & inTF2] # & inTF1 & inTF2

# movies[movies["Studio"] == "Universal"]
```

```python
# 제주 전입을 제일 많이 한 곳
JejuDF.sort_values(by="총계", ascending=False).head(1)[0]
```
