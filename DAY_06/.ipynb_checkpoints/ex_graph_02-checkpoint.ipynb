- 남북한 수력발전전력량에 대한 시각화

## <hr>(1) 모듈 로딩

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## <hr>(2) 데이터 준비

```python
file = "../DATA/남북한발전전력량.xlsx"
```

- 실제 데이터 파일 확인 결과
    * 첫 번째 행은 컬럼명 데이터
    * 첫 번째와 두 번재 컬럼을 행 인덱스 사용으로 적합
    * 입력값이 없는 쉘은 '-'채워져 있음 -> isnull로 안나오는 결측치임

## <hr>(3) 데이터 저장

```python
korpowerDF = pd.read_excel(file, index_col=[0,1])
```

## <hr>(4) 데이터 확인

### (4-1)데이터 타입과 실제 데이터 체크
* "-" -> 0으로 변경
* object -> uint16으로 변경

```python
korpowerDF.info()
```

### (4-2) 인덱스와 컬럼 속성 확인

```python
print(korpowerDF.index, korpowerDF.columns, sep="\n\n")
```

## <hr>(5) 데이터 전처리

### (5-1) 데이터 자료형 변환

```python
# "-" => "0"
korpowerDF2 = korpowerDF.replace("-", "0")
```

```python
# object = int16
korpowerDF2 = korpowerDF2.astype("int16")
```

## <hr>(6) 데이터 분석 작업
    -남북한의 수력발전량 비교하기

```python
korpowerDF2
```

```python
# 수력발전량 행 데이터 추출
waterDF = korpowerDF2.xs("수력", level=1)
```

```python
waterDF
```

```python
list(waterDF.columns)
```

```python
x = waterDF.columns
```

```python
# 남한
south = waterDF.loc["남한"]
```

```python
# 북한
north = waterDF.loc["북한"]
```

```python
plt.plot(x, south, x, north)
plt.show()
```
